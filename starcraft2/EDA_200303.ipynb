```python
import re
```

```python
def fix_missing(df, col):
    for i, row in df.iterrows():
        val = row[col]
        split = val.event_contents.split("-") # -로 나누면
        name = re.sub("[()]","",split[0]) # 빌드 건물이나 유닛 앞에 숫자만 가지고 올 수 있음
        col_name = "player" + str(val.player) + "_" + name
        col_name2 = "player" + str(val.player) + "_species"
        if col_name in df_train.columns:
            df_train[col_name] += 1
        else:
            df_train[col_name] = 1
            df_train[col_name2] = val.species
```

```python
fix_missing(df_train,["player","event_contents","species"])
```
